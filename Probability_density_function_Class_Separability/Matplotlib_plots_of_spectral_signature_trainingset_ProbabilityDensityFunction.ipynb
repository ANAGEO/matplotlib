## Create and export plots for a list of several features

The trick was found [here](https://stackoverflow.com/questions/20011494/plot-normal-distribution-with-matplotlib)

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Object feature to plot
feature_list=('ndvi_first_quart','ndwi_mean','opt_red_first_quart','perimeter','area')

for feature in feature_list:
    # Clear the figure, axes and close the existing figure
    plt.clf()
    plt.cla()
    plt.close()
    
    # Create a figure instance
    fig = plt.figure(1, figsize=(15, 15))

    nbrrowssubplot=4
    nbrcolumnssubplot=np.ceil((len(class_label)+1)/nbrrowssubplot)

    count=1
    for c in class_label: 

        tempdf=dfC.loc[dfC[class_column] == c].copy()
        data_C=np.around(tempdf.loc[:,[feature]][feature].tolist(),decimals=3)
        tempdf=dfA.loc[dfC[class_column] == c].copy()
        data_A=np.around(tempdf.loc[:,[feature]][feature].tolist(),decimals=3)

        plt.subplot(nbrrowssubplot,nbrcolumnssubplot,count)

        # Plot A curve
        data_A.sort()
        hmean = np.mean(data_A)
        hstd = np.std(data_A)
        pdf = stats.norm.pdf(data_A, hmean, hstd)
        plt.plot(data_A, pdf, c='green') # including h here is crucial

        # Plot A curve
        data_C.sort()
        hmean = np.mean(data_C)
        hstd = np.std(data_C)
        pdf = stats.norm.pdf(data_C, hmean, hstd)
        plt.plot(data_C, pdf, c='red') # including h here is crucial

        # Add a title for the current subplot
        plt.title(c)

        # Increment the count for the next subplot
        count+=1

    plt.subplots_adjust(top=0.92, bottom=0.08, left=0.10, right=0.95, hspace=0.30, wspace=0.15)
    
    # Save the figure
    output=os.path.join(output_folder,"Example_A_feature_"+str(feature)+".svg")
    fig.savefig(output, format='svg', dpi=800)
    
```

The trick to change colors on each loop was found [here](https://stackoverflow.com/questions/14720331/how-to-generate-random-colors-in-matplotlib).

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

def get_cmap(n, name='brg'):
    '''Returns a function that maps each index in 0, 1, ..., n-1 to a distinct 
    RGB color; the keyword argument name must be a standard mpl colormap name.'''
    return plt.cm.get_cmap(name, n)
```

```python
import numpy as np
import scipy.stats as stats
import matplotlib.pyplot as plt

# Generate a list of colors for each class
cmap = get_cmap(len(class_to_use))

# Object feature to plot
feature_list=('area','ndvi_first_quart')

for feature in feature_list:
    # Clear the figure, axes and close the figure
    plt.clf()
    plt.cla()
    plt.close()
    
    # Create a figure instance
    fig = plt.figure(1, figsize=(15, 15))

    for i, class_name in enumerate(class_to_use): 
        tempdf=dfC.loc[dfC[class_column] == class_name].copy()
        data_C=np.around(tempdf.loc[:,[feature]][feature].tolist(),decimals=3)
        tempdf=dfA.loc[dfC[class_column] == class_name].copy()
        data_A=np.around(tempdf.loc[:,[feature]][feature].tolist(),decimals=3)

        # Plot A curve
        data_A.sort()
        hmean = np.mean(data_A)
        hstd = np.std(data_A)
        pdf = stats.norm.pdf(data_A, hmean, hstd)
        plt.plot(data_A, pdf, linestyle='solid', c=cmap(i)) # including h here is crucial

        # Plot C curve
        data_C.sort()
        hmean = np.mean(data_C)
        hstd = np.std(data_C)
        pdf = stats.norm.pdf(data_C, hmean, hstd)
        plt.plot(data_C, pdf, linestyle='dashed', c=cmap(i)) # including h here is crucial
        
    # Save the figure
    output=os.path.join(output_folder,"Example_B_feature_"+str(feature)+".svg")
    fig.savefig(output, format='svg', dpi=800)
```
